````python
import matplotlib.pyplot as plt
import pickle
````

````python
merged = pickle.load(open('../code/experimental-notebooks/df_to_plot', 'rb'))
````

````python
merged.head()
````

````python
# set the value column that will be visualised
variable = 'Total Damage'
# set the range for the choropleth values
vmin, vmax = 0, 1000000000
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(55, 15))
# remove the axis
ax.axis('off')
# add a title and annotation
ax.set_title('Median %s per Registry'%variable, fontdict={'fontsize': '40', 'fontweight' : '3'})
ax.annotate('Source: Classification of Lexis Nexis Negligence Cases from 2000 to 2019', xy=(0.5, .05), xycoords='figure fraction', fontsize=20, color='#555555')
# Create colorbar legend
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm.set_array([]) # or alternatively sm._A = []. Not sure why this step is necessary, but many recommends it
# add the colorbar to the figure
fig.colorbar(sm)
# create map

# for idx, row in merged.iterrows():
#     ax.annotate(s=row['city'], xy=(row['lat'], row['lng']),horizontalalignment='center')

merged.plot(column = variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
````
